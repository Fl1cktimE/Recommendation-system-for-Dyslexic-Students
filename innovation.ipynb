```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as pit
%matplotlib inline
import matplotlib
matplotlib.rcParams["figure.figsize"] = (20,10)
```

```python
df1 = pd.read_csv("sheetdemo.csv")
df1.head()
```

```python
df1.shape
```

```python
df1['isNa'] = df1[df1.columns].isnull().apply(lambda x: all(x), axis=1)
df1.head()
```

```python
df1.drop(df1.loc[df1['isNa'] == True].index, inplace=True)
df1.head()
```

```python
df2 = df1.drop(['isNa'],axis='columns')
df2.head()
```

```python
df2['hearing problem'].fillna('No',inplace= True)
df2.head()
```

```python
df1['hearing problem'].fillna(df1['hearing problem'].value_counts().idxmax(),inplace= True)
df1.head()
```

```python
df3 = df1.drop(['isNa'],axis='columns')
df3.head()
```

```python
dummies = pd.get_dummies(df3.Age)
dummies.head(3)
```

```python
df3.columns = df3.columns.str.replace(' ', '')
df3.head()
```

```python
dummies = pd.get_dummies(df3.hearingproblem)
dummies.head()
```

```python
df4 = pd.concat([df3,dummies],axis='columns')
df4.head()
```

```python
df4.loc[df4['Visionproblem'] == 'yes','Visionproblem'] = 'yeah'
df4.head()
```

```python
df4.loc[df4['Visionproblem'] == 'no','Visionproblem'] = 'nope'
df4.head()
```

```python
dummies = pd.get_dummies(df4.Visionproblem)
dummies.head()
```

```python
df5 = pd.concat([df4,dummies],axis='columns')
df5.head()
```

```python
df6 = df5.drop(['Visionproblem','hearingproblem'],axis='columns')
df6.head()
```

```python
x = df6.drop(['writing','reading'],axis='columns')
x.head()
```

```python
y=df6.writing
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
len(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=2)
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
